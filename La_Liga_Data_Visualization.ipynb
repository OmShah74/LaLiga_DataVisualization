```python
#import libraries
import pandas as pd
import seaborn as sns
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data=pd.read_csv('combined_data_laliga.csv')
df=pd.DataFrame(data)
df
```

## DATA PREPROCESSING
Add a new column, score calculating the goal difference and delete all unnecessary columns.

```python
df1=df
df1['Score']=df1["Home Team Goals Scored"]-df1["Away Team Goals Scored"]
```

```python
df1 = df1.drop("Unnamed: 0", axis=1)
df
```

```python
df1 = df1.drop(["Half Time Score", "Home Team Off Target Shots", "Away Team Off Target Shots", "Home Team Blocked Shots", "Away Team Blocked Shots", "Home Team Corners", "Away Team Corners", "Home Team Throw Ins", "Away Team Throw Ins", "Home Team Aerials Won", "Away Team Aerials Won", "Home Team Clearances", "Away Team Clearances", "Home Team Second Yellow Cards", "Away Team Second Yellow Cards"], axis=1)
```

```python
#Create new column for total shots and for total fouls
df1["Total Shots"] = df1["Home Team Total Shots"] + df1["Away Team Total Shots"]
df1["Total Fouls"] = df1["Home Team Fouls"] + df1["Away Team Fouls"]
```

```python
#Creating Winning Team Column -> 1 is when home team wins, -1 is when away team wins, 0 is draw
def winning_team(score):
  if score > 0:
    return 1
  elif score == 0:
    return 0
  else:
    return -1
```

```python
df1["Winning Team"] = df1.Score.apply(lambda x:winning_team(x))
df1
```

## Data Visualization

```python
#Heatmap
corr = df1.corr()
plt.figure(figsize=(10,10))
sns.heatmap(corr)
plt.xticks(rotation=70)
```

```python
plt.bar(df1['year'],df1['Home Team Goals Scored'],color='orange')
plt.title('Year vs Goals for Home Team')
plt.xlabel('Year')
plt.ylabel('Goals')
plt.show()
```

```python
plt.bar(df1['year'],df1['Away Team Goals Scored'],color='purple')
plt.title('Year vs Goals for Away Team')
plt.xlabel('Year')
plt.ylabel('Goals')
plt.show()
```

## Correlation Analysis

```python
# show correlation between Total shots and Total fouls with Match excitement
plt.figure(figsize=(10,5))
ax = plt.subplot(1,2,1)
sns.scatterplot(x="Total Shots", y="Match Excitement", data=df1, ax=ax)
ax = plt.subplot(1,2,2)
sns.scatterplot(x="Total Fouls", y="Match Excitement", data=df1, ax=ax)
```

 There is a clear positive correlation between the total number of shots and the match excitement rating. However, this is not the same case between total fouls and match excitement. So we can infer that people get excited by goals and are not influenced by fouls.

```python
#Create new column for total goals
df1["Total Goals"] = df1["Home Team Goals Scored"] + df1["Away Team Goals Scored"]
```

```python
#Correlation between (Match Excitement, Total Goals)
plt.figure(figsize=(10,5))
sns.scatterplot(x="Total Goals", y="Match Excitement", data=df1)
```

There is a clear positive correlation between the total number of goals and the match excitement rating. However, there is 1 outlier that has a relatively low rating but high amount of goals, might be a data anamoly.

```python
#Correlation between (Winning Team Rating, Winning Team Goals)
#For this, we will create 2 new DFs, (Winning Team (home), Home Team Rating, Home Goals) and (Winning Team (away), Away Team Rating, Away Goals)
home_df = df1[:].copy()
home_df = home_df[home_df["Winning Team"] > 0]
home_df['Winning Team'].value_counts()
```

```python
away_df = df1[:].copy()
away_df = away_df[away_df["Winning Team"] < 0]
away_df['Winning Team'].value_counts()
```

```python
plt.figure(figsize=(10,5))
ax = plt.subplot(1,2,1)
sns.scatterplot(x="Home Team Goals Scored", y="Home Team Rating", data=home_df, ax=ax)
ax = plt.subplot(1,2,2)
sns.scatterplot(x="Away Team Goals Scored", y="Away Team Rating", data=away_df, ax=ax)
```

There is a clear positive correlation between the total number of goals and the match excitement rating. However, there is 1 outlier that has a relatively low rating but high amount of goals, might be a data anamoly.

```python
#Correlation between (Winning Team Rating, Winning Team Goals)
#For this, we will create 2 new DFs, (Winning Team (home), Home Team Rating, Home Goals) and (Winning Team (away), Away Team Rating, Away Goals)
home_df = df1[:].copy()
home_df = home_df[home_df["Winning Team"] > 0]
home_df['Winning Team'].value_counts()
```

```python
away_df = away_df[away_df["Winning Team"] < 0]
away_df['Winning Team'].value_counts()
```

```python
plt.figure(figsize=(10,5))
ax = plt.subplot(1,2,1)
sns.scatterplot(x="Home Team Goals Scored", y="Home Team Rating", data=home_df, ax=ax)
ax = plt.subplot(1,2,2)
sns.scatterplot(x="Away Team Goals Scored", y="Away Team Rating", data=away_df, ax=ax)
```

There is a clear positive correlation between the Winning Team Rating and Number of goals scored by the winning team.

```python
#Correlation between (Total Fouls, Total Yellow / Red Cards)
#For this, we will create 2 new columns to calculate the total fouls and total cards
df1["Total Fouls"] = df1['Away Team Fouls'] + df1['Home Team Fouls']
df1["Total Cards"] = df1['Home Team Red Cards'] + df1['Home Team Yellow Cards'] + df1['Away Team Yellow Cards'] + df1['Away Team Red Cards']
```

```python
plt.figure(figsize=(10,5))
sns.lineplot(x="Total Fouls", y="Total Cards", data=df1)
```

There is a clear positive correlation between the number of fouls and the total number of cards.

```python
#Correlation between (Winning Team Pass Success, Winning Team Total Shots)
plt.figure(figsize=(10,5))
ax = plt.subplot(1,2,1)
sns.lineplot(y="Home Team Pass Success %", x="Home Team Total Shots", data=home_df, ax=ax)
ax = plt.subplot(1,2,2)
sns.lineplot(y="Away Team Pass Success %", x="Away Team Total Shots", data=away_df, ax=ax)
```

Both the plots prove that even if there is a high success rate of passing, we can't expect a higher amount of shots. However, there is an important point to note here which is that for a team to shoot more, they must have a high % of passing success rate.

```python
#Correlation between (Year, Total Shots for one team)
#Take Real Madrid and put it in 2 new Data Frames (home and away)
madridhome_df = df1.copy()
madridaway_df = df1.copy()
madridhome_df = madridhome_df[madridhome_df['Home Team'] == 'REAL MADRID']
madridaway_df = madridaway_df[madridaway_df['Away Team'] == 'REAL MADRID']
plt.figure(figsize=(10,5))
ax=plt.subplot(1,2,1)
sns.scatterplot(x="year", y="Home Team Goals Scored", data=madridhome_df, ax=ax)
ax=plt.subplot(1,2,2)
sns.scatterplot(x="year", y="Away Team Goals Scored", data=madridaway_df, ax=ax)
```

 There is no clear correlation between the number of goals scored and the years.

## Data Analysis

```python
#Do teams that play at home score more goals?
#For this, we create 2 new data frames for the home team and the away team
hometeam_df = df1.filter(["Home Team", "Home Team Goals Scored"])
hometeam_df.groupby("Home Team").mean().sum()
```

```python
awayteam_df = df1.filter(["Away Team", "Away Team Goals Scored"])
awayteam_df.groupby("Away Team").mean().sum()
```

```python
plt.figure(figsize=(50,10))
ax = plt.subplot(1,2,1)
sns.barplot(y="Home Team", x="Home Team Goals Scored", data=home_df, ax=ax)
ax = plt.subplot(1,2,2)
sns.barplot(y="Away Team", x="Away Team Goals Scored", data=away_df, ax=ax)
```

On an average, teams that play at home score more goals.

```python
#Do Home Teams concede less goals?
hometeam_df = df1.filter(["Home Team", "Home Team Goals Conceeded"])
hometeam_df.groupby("Home Team").mean().sum()
```

```python
awayteam_df = df1.filter(["Away Team", "Away Team Goals Conceeded"])
awayteam_df.groupby("Away Team").mean().sum()
```

```python
plt.figure(figsize=(50,10))
ax = plt.subplot(1,2,1)
sns.barplot(y="Home Team", x="Home Team Goals Conceeded", data=home_df, ax=ax)
ax = plt.subplot(1,2,2)
sns.barplot(y="Away Team", x="Away Team Goals Conceeded", data=away_df, ax=ax)
```

On an average, teams that play at home conceeded lesser goals.
